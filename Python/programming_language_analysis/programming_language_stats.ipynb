```python
# swap boolean result
filtered_df = filtered_df.loc[~filtered_df.duplicated(subset=['language_name', 'category', 'last_activity_year'])].reset_index(drop=True).copy()
```

```python
# check datatypes according to user input
def find_columns_of_types(df, type):
    return df.columns[df.dtypes == type].tolist()

# invalid characters remove function
def remove_invalid_characters(df, columns=[]):
    pattern = r'[^a-zA-Z0-9\s]'
    temp = df
    for column in columns:
        temp[column] = temp[column].astype(str).str.strip().replace(pattern, '', regex=True)
    return temp

# get all object columns
object_list = find_columns_of_types(filtered_df, 'object')
# exclude creators from the list
list_to_remove = ['language_name', 'category', 'creator']
object_list = list(set(object_list) - set(list_to_remove))
filtered_df = remove_invalid_characters(filtered_df, object_list)

# title for indeed_jobs_title
filtered_df['indeed_jobs_title'] = filtered_df['indeed_jobs_title'].str.title()
```

```python
# create a function for generating a scatter with preconfigured parameters
def scatterplot_generator(df, xcol, ycol, title, xlabel, ylabel, figsize=(6.5, 6.5), size=""):
    # configure the plt settings
    f, ax = plt.subplots(figsize=figsize)
    sns.despine(f, left=True, bottom=True)
    # resolve scientific notation display
    plt.ticklabel_format(style='plain', axis='both')

    # Scatter plot 
    sc_plt = sns.scatterplot(df, x=xcol, y=ycol, size=size, ax=ax)
    # Set title and labels
    sc_plt.set_title(label=title)
    sc_plt.set(xlabel=xlabel, ylabel=ylabel)
    return sc_plt
```

```python
# Find relationship between total user of the language and the linkedin skills in 2018
tul_ls_plt = scatterplot_generator(vis_data, "total_jobs_in_market", "linkedin_skill_2018", "Total User vs LinkedIn Skill in 2018 - Positive Correlation", "Total_User", "LinkedIn Skill in 2018", size="linkedin_skill_2018")
```

```python
# Filter the top 10 indeed jobs in 2017
lti_df = vis_pl_df_9023.loc[~vis_pl_df_9023['indeed_jobs_2017'].isna()][['language_name', 'total_jobs_in_market', 'indeed_jobs_2017']].sort_values(by=['total_jobs_in_market', 'indeed_jobs_2017'], ascending=[False, False]).head(10).copy()

# define proportion column conditions from Major to Minor
lti_df['proportion'] = lti_df['indeed_jobs_2017'].apply(lambda x: 'Major' if x >= 30000 else ('Above Average' if x >= 20000 and x < 30000 else ('Average' if x >= 10000 and x < 20000 else 'Minor')))

lti_df['proportion_idx'] = lti_df['proportion'].apply(lambda x: 1 if x == 'Major' else (2 if x == 'Above Average' else (3 if x == 'Average' else 4)))

lti_df = lti_df.sort_values(by="proportion_idx", ascending=False)
```

```python
f, ax = plt.subplots(figsize=(10, 10))
sns.despine(f)
top10_indeed_2017_hplt = sns.histplot(lti_df, x="indeed_jobs_2017", hue="proportion", multiple="stack", palette='viridis', ax=ax)

top10_indeed_2017_hplt.set_title(label="Top 10 Indeed Jobs in 2017 Proportion by Programming Language - HistPlot")
top10_indeed_2017_hplt.set(xlabel=" Programming Language")
```

```python
# Top 10 Indeed Jobs in 2017 Proportion by Programming Language
f, ax = plt.subplots(figsize=(10, 10))
sns.despine(f)
top10_indeed_2017_bplt = sns.barplot(lti_df, x="language_name", y="indeed_jobs_2017", hue="proportion", palette="viridis", ax=ax)

top10_indeed_2017_bplt.set_title(label="Top 10 Indeed Jobs in 2017 Proportion by Programming Language - Barplot")
top10_indeed_2017_bplt.set(xlabel=" Programming Language", ylabel="Indeed Jobs in 2017")
```
