```python
# swap boolean result
filtered_df = filtered_df.loc[~filtered_df.duplicated(subset=['language_name', 'category', 'last_activity_year'])].reset_index(drop=True).copy()
```

```python
# check datatypes according to user input
def find_columns_of_types(df, type):
    return df.columns[df.dtypes == type].tolist()

# invalid characters remove function
def remove_invalid_characters(df, columns=[]):
    pattern = r'[^a-zA-Z0-9\s]'
    temp = df
    for column in columns:
        temp[column] = temp[column].astype(str).str.strip().replace(pattern, '', regex=True)
    return temp

# get all object columns
object_list = find_columns_of_types(filtered_df, 'object')
# exclude creators from the list
list_to_remove = ['language_name', 'category', 'creator']
object_list = list(set(object_list) - set(list_to_remove))
filtered_df = remove_invalid_characters(filtered_df, object_list)

# title for indeed_jobs_title
filtered_df['indeed_jobs_title'] = filtered_df['indeed_jobs_title'].str.title()
```

```python
# change year columns to int64
year_type = {
    'wiki_first_appeard_year': int,
    'wiki_page_created_year': int,
    'githubrepo_update_year': int,
    'githubrepo_created': int
}

year_field = [
    'wiki_first_appeard_year',
    'wiki_page_created_year',
    'githubrepo_update_year',
    'githubrepo_created'
    ]


filtered_df[year_field] = filtered_df[year_field].fillna(0)
# filtered_df['wiki_page_created_year'] = filtered_df['wiki_page_created_year'].fillna(0)
# filtered_df['githubrepo_update_year'] = filtered_df['githubrepo_update_year'].fillna(0)
# filtered_df['githubrepo_created'] = filtered_df['githubrepo_created'].fillna(0)

filtered_df = filtered_df.astype(year_type)
```

```python
# create a new df that includes dataset from 1990 - 2023
vis_pl_df_9023 = vis_data.loc[(vis_data['publish_year'] >= 1990) & (vis_data['category'] == 'pl')].copy()

gopl = vis_pl_df_9023['publish_year'].value_counts().reset_index()

gopl
```

```python
plt.figure(figsize=(19, 8))
# LinePlot - number of programming language growth from 1990 - current
goplby_plt = sns.lineplot(gopl, x="publish_year", y="count")
# define attributes for the figure
goplby_plt.set_title(label="Growth of Programming Languages from 1990 - 2022")
# configure the figure size and xticks
plt.xticks(ticks=gopl['publish_year'], labels=gopl['publish_year'])
goplby_plt.set(xlabel='Year', ylabel='Number of Programming Laguage Grow')
```

```python
# last activity year of programming language by number of total users from 1990 - 2022
mppl_df = vis_pl_df_9023

# find 
# mppl_df.loc[(mppl_df['publish_year'] == 1992)].sort_values(by='total_user', ascending=False)
idx = mppl_df.groupby('last_activity_year')['total_user'].idxmax()
lapl_df = mppl_df.loc[idx].reset_index(drop=True).copy()

lapl_df[['language_name', 'last_activity_year', 'total_user']]
```

```python
# filter the most popular programming languages from df (e.g. Java, Javascript, Python, C++, C#) by stackoverflow_fans_votes
top10_stackoverflow_pl_by_fans = vis_pl_df_9023.sort_values(by="stackoverflowsurvey_fans_2021", ascending=False).head(10)['language_name'].tolist()

top10_stackoverflow_pl_by_fans
```

```python
# top10_stackoverflow votes programming languages by fans
top10_df = vis_pl_df_9023.loc[(vis_pl_df_9023['language_name'].isin(top10_stackoverflow_pl_by_fans))].copy()

top10_df = top10_df.sort_values(['stackoverflowsurvey_fans_2021']).reset_index(drop=True)
# set plt figure
plt.figure(figsize=(12, 8))
# Barplot
top10_sovv_plt = sns.barplot(top10_df, x="language_name", y="stackoverflowsurvey_fans_2021", hue="language_name")
top10_sovv_plt.set_title(label="Top 10 Most Popular Programming Languages Voted by Stackoverflow Users in 2021")
top10_sovv_plt.set(xlabel="Programming Languages", ylabel="Stackoverflow fans Vote")
```

```python
# Top 10 Use of Programming Languages by total user
top15_uopl_df = vis_pl_df_9023.sort_values(by="total_user", ascending=False).head(10).copy()

plt.figure(figsize=(18,8))
uopl_plt = sns.barplot(top15_uopl_df, x="language_name", y="total_user", hue="language_name")
plt.ticklabel_format(style='plain', axis='y')
uopl_plt.set_title("Top 10 Use of Programming Languages by Total User")
uopl_plt.set(xlabel="Programming Language", ylabel="Total User")
```

```python
# Top 15 highest median salary programming language users in 2021
top15_hmspl_df = vis_pl_df_9023.sort_values(by=['stackoverflowsurvey_mediansalary_2021'], ascending=False).head(15).copy()

f, ax = plt.subplots(figsize=(10, 10))
sns.despine(f)

# Barplot
hmspl_plt = sns.barplot(top15_hmspl_df, x='stackoverflowsurvey_mediansalary_2021', y='language_name', hue='language_name',ax=ax)

hmspl_plt.set_title(label="Top 15 Highest Median Salary Programming Language Users in 2021")
hmspl_plt.set(xlabel="Median Salaray", ylabel="Programming Languagh")

```
